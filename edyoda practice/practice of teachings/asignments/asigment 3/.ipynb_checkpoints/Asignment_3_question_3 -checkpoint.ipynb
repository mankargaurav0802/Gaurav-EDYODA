```python
def count_cases1(sample):
    upper_case_in_str = 0
    lower_case_in_str = 0
    for count in sample:
        if count.isupper():
            upper_case_in_str += 1
        elif count.islower():
            lower_case_in_str += 1
    return "No. of Upper case characters : {}  No. of Lower case Characters : {}".format(upper_case_in_str,lower_case_in_str)
```

```python
Sample_String='The quick Brow Fox'
```

```python
count_cases1(Sample_String)
```

```python
sample= eval(input())
count_cases1(sample)
```
